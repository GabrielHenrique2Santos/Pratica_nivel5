```python
!pip install faker
!pip install wordcloud
!pip install num2words

from faker import Faker
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from num2words import num2words
```

```python
gera = Faker()
aluno = [[gera.name(),gera.random_int(0,10)] for x in range(10)]
arquivo = open('arquivo.txt','w+')
[arquivo.write(f"{x}\n") for x in aluno]
arquivo.close()
print(aluno)
```

```python
lerArquivo = open('arquivo.txt')
a = lerArquivo.readlines()
dados = []
for b in a:
  for c in b:
    if c.isdigit() == True:
      dados.append(int(c))

linha, barra, coluna = plt.hist(dados,facecolor='blue',alpha=0.75); 

plt.xlabel('Notas')
plt.ylabel('Chance')
plt.title('Histograma das notas')
plt.grid(True)
plt.show()
```

```python
nuvem = []

for numeros in dados:
  numero = num2words(numeros, lang='pt_br')
  nuvem.append(numero)

nuvem = ",".join(nuvem)
nuvem_palavras = WordCloud(background_color = 'black', width=800, height=400).generate(nuvem)
plt.imshow(nuvem_palavras, interpolation = 'bilinear')
plt.axis("off")
nuvem_palavras.to_file("nuvemDePalavras.png")
```

# **Análise e Conclusão:**

**Qual a importância de manipular arquivos de texto?**

A manipulação de dados em arquivos de texto é de suma importancia para prevenir a necessidade de criar um novo arquivo do zero para cada modificação, tornando assim possivel a atualização de arquivos ja existentes, criado uma cadeia de progressão ao inves de novos arquivos, tornando assim possivel a existencia de bancos de dados, listas, catalogos e etc.

**Qual a importância de visualizar os dados em um histograma?**

Um histograma facilita na visualização de dados em massa, reduzindo consideravelmente o tamanho das informações a serem analisadas para chegar a uma mesma conclusão, tornando o processo de analize mais comodo e pratico.

**O que significa a visualização por nuvem de palavras?**

Visualizar por nuvem de palavras é mantes um destaque para as, nesse caso, palavras que aparecem com mais frequencia no histograma, novamente facilitando na compreensão atravez de um instrumento grafico.
